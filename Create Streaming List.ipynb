```python
# Import Packages
from googleapiclient.discovery import build
import datetime as dt
from bs4 import BeautifulSoup
import requests
import networkx as nx
import matplotlib.pyplot as plt
from pyvis.network import Network
import pickle
import difflib
import re
import string
```

```python
# Will parse reelgood website to get names of anime in given streaming service
def get_stream_service(url, offset):
    resp = requests.get(url)
    #search_term = ".css-1u7zfla a"
    search_term = ".e1qyeclq4 p"
    # get BeautifulSoup object
    soup = BeautifulSoup(resp.content, 'html.parser')
    # Get the titles of the user recommended animes
    links = soup.select(search_term)
    recs = [title.text for title in links]
    for i in range(50,offset+1,50):
        offset_url = url + "?offset=" + str(i)
        #print(offset_url)
        resp = requests.get(offset_url)
        # get BeautifulSoup object
        soup = BeautifulSoup(resp.content, 'html.parser')
        # Get the titles of the user recommended animes
        links = soup.select(search_term)
        recs.extend([title.text for title in links])
    return recs

# This will find what streaming platforms the given anime is on, based on reelgood website
# Will also assign year of anime based on reelgood
def get_stream(anime_name):
    stream_list = []
    if(anime_name in crunch_list):
        stream_list.append("Crunchyroll")
    if(anime_name in netflix_list):
        stream_list.append("Netflix")
    if(anime_name in prime_list):
        stream_list.append("Amazon Prime")
    if(anime_name in funimation_list):
        stream_list.append("Funimation")
    if(anime_name in hbo_list):
        stream_list.append("HBO")
    if(anime_name in hulu_list):
        stream_list.append("Hulu")
    #year = re.findall(r"\((\d+)\)", anime_name)
    #stream_year = [stream_list,year]
    return stream_list
def pickle_list(df, file_name):
    with open(file_name, 'wb') as handle:
        pickle.dump(df, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
stream_list = {}
crunch_list = get_stream_service("https://reelgood.com/source/crunchyroll", 600)
netflix_list = get_stream_service("https://reelgood.com/genre/anime/on-netflix", 100)
prime_list = get_stream_service("https://reelgood.com/genre/anime/on-amazon", 50)
funimation_list = get_stream_service("https://reelgood.com/source/funimation", 550)
hbo_list = get_stream_service("https://reelgood.com/genre/anime/on-hbo_max", 50)
hulu_list = get_stream_service("https://reelgood.com/genre/anime/on-hulu", 200)
```

```python
full_anime_list = set(crunch_list + netflix_list + prime_list + funimation_list + hbo_list + hulu_list)
```

```python
stream_list = {re.findall(r"(.*)\s\(\d+\)", anime)[0]: get_stream(anime) for anime in full_anime_list}
```
